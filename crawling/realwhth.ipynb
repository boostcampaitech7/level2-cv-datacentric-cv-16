```python
# 각 nth-child 위치에 있는 a 태그의 href 속성 가져오기
href_list = []
for i in range(1, 44):  # 1부터 43까지 반복
    try: 
        # nth-child 위치에 있는 a 태그의 href 선택
        css_selector = f"#rso > div > div > div.wH6SXe.u32vCb > div > div > div:nth-child({i}) > div.czzyk.XOEbc > h3 > a"
        link = driver.find_element(By.CSS_SELECTOR, css_selector)
        href = link.get_attribute("href")
        print(href)
        if href:
            href_list.append(href)
            print(f"{i}번째 href: {href}")
    except Exception as e:
        print(f"{i}번째 요소를 찾지 못했습니다: {e}")

# 브라우저 닫기
# driver.quit()

# 결과 확인
print("모든 href 링크:", href_list)
```

```python
image_list = []
for href in href_list:
    driver.get(href)  # URL로 이동
    time.sleep(2)
    css_selector = f"#imp > div:nth-child(1) > div.A8mJGd.NDuZHe > div.LrPjRb > div > div.BIB1wf.EIehLd.fHE6De.Emjfjd > c-wiz > div > div.v6bUne > div > a > img.sFlh5c.FyHeAf.iPVvYb"
    image = driver.find_element(By.CSS_SELECTOR, css_selector)
    image_src = image.get_attribute("src")
    image_list.append(image_src)
    print(image_src)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
programs = soup.select('#imp > div:nth-child(1) > div.A8mJGd.NDuZHe > div.LrPjRb > div > div.BIB1wf.EIehLd.fHE6De.Emjfjd > c-wiz > div > div.v6bUne > div > a > img.sFlh5c.FyHeAf.iPVvYb') 
src_list = []
# 각 요소에 대해 반복
for program in programs:
    if program.has_attr('src'):
        src_list.append(program['src'])  # src 속성 값 추가
        print(program['src'])
```

```python
import requests
import base64
import re
import os

# 저장 폴더 생성
os.makedirs("images", exist_ok=True)

for idx, src in enumerate(image_list):
    # Base64 인코딩된 이미지인지 확인
    if src.startswith("data:image"):
        # Base64 데이터를 추출하여 디코딩
        base64_data = re.sub('^data:image/.+;base64,', '', src)
        image_data = base64.b64decode(base64_data)
        
        # 이미지 파일로 저장
        with open(f"Thailand_receipts/image_{idx}.jpg", "wb") as f:
            f.write(image_data)
        print(f"Base64 이미지 {idx} 저장 완료")
    
    # URL로 직접 접근 가능한 이미지인 경우
    else:
        response = requests.get(src)
        
        # 성공적으로 접근했는지 확인
        if response.status_code == 200:
            with open(f"Thailand_receipts/image_{idx}.jpg", "wb") as f:
                f.write(response.content)
            print(f"URL 이미지 {idx} 저장 완료")
        else:
            print(response.status_code)
            print(f"이미지 {idx} 다운로드 실패")
```
