```python
import json
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib import patches
```

```python
where = 'img/test'
nation_dict = {
    'vi': 'vietnamese_receipt',
    'th': 'thai_receipt',
    'zh': 'chinese_receipt',
    'ja': 'japanese_receipt',
}
```

```python
with open('predictions/base150epoch_test.csv') as f:
    json_data = json.load(f)
json_data['images'].keys()

img_names=sorted(list(json_data['images'].keys()))

# Image.open(img_names[i]) 쓰면 바로 이미지
for img_num in range(len(img_names)): 
    nation = img_names[img_num].split('.')[1]
    img_names[img_num] = './data/'+nation_dict[nation]+'/'+where+'/'+img_names[img_num]
```

```python
i=0
```

```python
words = json_data['images'][img_names[i].split('/')[-1]]['words']
fig, ax = plt.subplots(1,2,figsize=[12,7])
ax[0].imshow(Image.open(img_names[i]))
ax[1].imshow(Image.open(img_names[i]))

for num in words:
    poly = patches.Polygon(words[num]['points'], linewidth=1, edgecolor='r', facecolor='none')
    ax[0].add_patch(poly)
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].set_xticks([])
ax[1].set_yticks([])
plt.show()
i += 1
```
