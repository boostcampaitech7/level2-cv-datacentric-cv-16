```python
import os
import json
from glob import glob
from pathlib import Path
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
want = 'ja'
nation_dict = {
    'vi': 'vietnamese_receipt',
    'th': 'thai_receipt',
    'zh': 'chinese_receipt',
    'ja': 'japanese_receipt',
}
```

```python
with open(os.path.join(os.path.join('./code/data', nation_dict[want]),'ufo/train.json')) as f:
    json_data=json.load(f)['images']
# print(json_data.keys())
img_names = list(json_data.keys())
# len(img_names)
```

```python
i=0
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,7))
print(img_names[i])
ax[0].imshow(Image.open(os.path.join(os.path.join(os.path.join('./code/data', nation_dict[want]), 'img/train'), img_names[i])))
ax[1].imshow(Image.open(os.path.join(os.path.join(os.path.join('./code/data', nation_dict[want]), 'img/train'), img_names[i])))
for num in json_data[img_names[i]]['words'].keys():
    points = json_data[img_names[i]]['words'][num]['points']
    x1, y1 = points[0]
    x2, y2 = points[2]
    rect = patches.Rectangle((x1, y1), x2-x1, y2-y1, linewidth=1, edgecolor='r', facecolor='none')
    ax[0].add_patch(rect)
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].set_xticks([])
ax[1].set_yticks([])
plt.show()
i+=1
```
